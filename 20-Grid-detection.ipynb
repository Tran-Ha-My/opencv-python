# GRID DETECTION

# Usage

Oftentimes when tracking object you might see some **Camera Distortion**. 

To solve this, when performing object tracking, it's good to have a **_recognizable pattern attached to the object tracked_**

=> Use **Grid patterns** to calibrate cameras and track motion

## Built-in functions


* **cv2.findChessboardCorners(src, (m,n))**

* **cv2.drawChessboardCorners(src, (m,n), corners, found)**

* **cv2.findCirclesGrid(src, (m,n), cv2.CALIB_CB_SYMMETRIC_GRID)**

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
found, corners = cv2.findChessboardCorners(flat_chess, (7,7)) #note that thou the board is 8x8, the last two aren't going to be detected so it's only (7,7)
```

```python
found
```

```python
cv2.drawChessboardCorners(flat_chess, (7,7), corners, found) # these 2 functions works together nicely
```

```python
plt.imshow(flat_chess)
```

```python
dots = cv2.imread('DATA/dot_grid.png')
```

```python
plt.imshow(dots)
```

```python
found, corners = cv2.findCirclesGrid(dots, (10,10), cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
found
```

```python
cv2.drawChessboardCorners(dots, (10,10), corners, found) 
plt.imshow(dots)
```
