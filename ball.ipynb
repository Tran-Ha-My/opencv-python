```python
def show_pic (img):
    fig = plt.figure( figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
show_pic(img)
```

```python
#### TASK: anywhere there's ink => black, else => white  => just black and white no grey

ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
show_pic(thresh1)
```

```python
## THE IMAGE LOSES QUALITY bc grey color is turned to white ! PLEASE FIX ###
```

```python
### ADAPTIVE THRESHOLD ###
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 8)
show_pic(thresh2)
```

I actually switch the 4th parameter to THRESH_BINARY_INV method instead of just binary because I want the black background and white edges 

```python
####BLENDING RESULTS TO FILL IN THE WHITE SPACES IN THRESH2####
blending = cv2.addWeighted(src1 = thresh1, src2 = thresh2, alpha = 0.7, beta = 0.4,gamma = 0)
show_pic(blending)
```
